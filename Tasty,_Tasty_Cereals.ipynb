In this part we will explore a data-set of cereals in the file [`Edible_grains.csv`](https://drive.google.com/file/d/177adJlaUWDL1-s97lbjTPshdpRJ7_Lr2/view?usp=sharing). The original file and description can be found on

E1. Creating a DataFrame and load the data. Displaying the first 5 and last 5 rows of the DataFrame.

```python
import pandas as pd 
df = pd.read_csv('Edible_grains.csv') 
df.head(5)
```

```python
df.tail(5)
```

E2. Using the `drop` function to delete the following columns from the data-set: `shelf`, `weight` and `cups`

```python
grouped = df.groupby('grain_name').sum()
grouped.drop(['shelf', 'weight', 'cups'], axis=1, inplace=True)
grouped
```

E3. Suppose we are interested in getting a sense of the relationship among several of the fields. Creating a scatterplot matrix of the following fields: `calories_content`, `protein_content`, `fat_content`, `sodium_content`, `fiber_content` and `sugars_content`.

```python
splom = pd.plotting.scatter_matrix(df[['calories_content', 'protein_content','fat_content', 'sodium_content', 'fiber_content', 'sugars_content']], alpha=0.4, figsize=(12, 12))  
```

E4. Using the `corr` fuction to generate a correlation matrix of the  data-set. pandas will take care of excluding the non-numerical values.

```python
df.corr()
```

E7. Suppose we are on the hunt for a cereal that is high in fiber, low in sodum, and high in protein. Creating a scatterplot with `fiber_content` for the x-axis, `sodium_content` for the y-axis, and color the points using `protein_content`. Using the `RdYlGn` colormap. 

```python
df.plot.scatter(x='fiber_content', y='sodium_content', c='protein_content', colormap='RdYlGn', sharex=False)
```
